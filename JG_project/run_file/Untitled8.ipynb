```python
import cv2
classes_path = './valid/_annotations.txt'
with open(classes_path) as f:
    class_names = f.readlines()
    for x in class_names:
        l = x.split(' ')
        l[-1] = l[-1].rstrip('\n')
        img_path = './valid/'+ l[0]
        img = cv2.imread(img_path)
        for i in l[1:]:
            m = i.split(',')
            m[0] = int(m[0])
            m[1] = int(m[1])
            m[2] = int(m[2])
            m[3] = int(m[3])
            m[4] = int(m[4])
            if(m[4] == 1):
                cv2.rectangle(img, (m[0],m[1]), (m[2],m[3]), (0,0,255), 2)
#                 cv2.rectangle(img, (m[0],m[1]), (3,3), (0,0,255), thickness=2)
                cv2.putText(img, "bus", (m[0],m[1]-2), cv2.FONT_HERSHEY_SIMPLEX, 2/3, (0,0,255), 1)
            if(m[4] == 0):
                cv2.rectangle(img, (m[0],m[1]), (m[2],m[3]), (0,255,0), 2)
#                 cv2.rectangle(img, (m[0],m[1]), (3,3), (0,0,255), thickness=2)
                cv2.putText(img, "car", (m[0],m[1]-2), cv2.FONT_HERSHEY_SIMPLEX, 2/3, (0,255,0), 1)
            if(m[4] == 2):
                cv2.rectangle(img, (m[0],m[1]), (m[2],m[3]), (255,0,0), 2)
#                 cv2.rectangle(img, (m[0],m[1]), (3,3), (0,0,255), thickness=2)
                cv2.putText(img, "truck", (m[0],m[1]-2), cv2.FONT_HERSHEY_SIMPLEX, 2/3, (255,0,0), 1)
        cv2.imwrite("./valid/rec/"+l[0],img)
#         cv2.imshow("tets", img)
cv2.waitKey(0)  #키보드 눌림 대기
cv2.destroyAllWindows();
```
